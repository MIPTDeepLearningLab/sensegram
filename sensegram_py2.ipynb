# WSD

```python
from pymystem3 import Mystem
import pandas as pd
stemmer = Mystem()
```

```python
from gensim.models import word2vec
wv = word2vec.Word2Vec.load_word2vec_format("./model/main_words_wiki_normalized_no_punct.txt.words",\
                                            binary=True)
```

```python
wsd_model = sensegram.WSD(sv, wv, window=5, method='sim', filter_ctx=3)
```

```python
t1 = u"бор ( элемент ) бор — элемент тринадцатой группы ( по устаревшей классификации — главной подгруппы третьей группы ) , второго периода периодической системы химических элементов с атомным номером . обозначается символом b ( ) . в свободном состоянии"
t2 = u"сосновый бор ( город ) сосно́вый бор — город ( с года ) в ленинградской области россии , образует муниципальное образование сосновоборский городской округ . ранее на месте города располагалась старинная деревня устья . впервые деревня"
```

```python
wsd_model.dis_text(t1, "бор", 0, 3)
```

```python
wsd_model.dis_text(t2, "бор", 131, 134)
```

# Test system

```python
wiki_tr = pd.read_csv("/home/fogside/Projects/russe_wsi/data/main/wiki-wiki/train.csv", sep='\t')
wiki_tr.head()
```

```python
wiki_tr[(wiki_tr.word == "бор")]
```

```python
labels = []
w = u"бор"

posit = wiki_tr[wiki_tr.word == w].positions
positions =[]
for pos in posit:
     positions.append([p.split('-') for p in pos.split(',')])

contexts = wiki_tr[wiki_tr.word == w].context
contexts = [" ".join(stemmer.lemmatize(c)) for c in contexts]

for c,p in zip(contexts,positions):
    print(p)
    p1 = p[0][0]
    p2 = p[0][1]
    mark = wsd_model.dis_text(unicode(c), w, int(p1),int(p2))
    if mark is not None:
        mark = mark[0]
    labels.append(mark)
    
```

```python
labels
```

```python
with open("kek_casle_output.txt", 'w') as fw:
    for l in labels:
        fw.write(str(l)+'\n')
```
